# Binary Classification

**Binary classification** is a fundamental concept in machine learning and statistics. It is a type of supervised learning task in which the goal is to categorize data points into one of two possible classes or categories. In binary classification, there are two mutually exclusive and exhaustive classes: the positive class (often labeled as "1" or "positive") and the negative class (labeled as "0" or "negative"). The objective is to assign each data point to one of these two classes based on its features or attributes.

Depond on the application, choosing a decision threshold for classifying data points into the positive or negative class is one of the most important thing that you should do. This threshold can be adjusted to achieve a desired trade-off between precision and recall.


## Introduction to Naive Bayes algorithm

Explain Naive Bayes algorithm and the "naive" assumption in Naive Bayes

The Naive Bayes algorithm is a family of probabilistic algorithms that apply Bayes' theorem with a strong (naive) independence assumption between the features. It's widely used in classification tasks because it's simple, easy to implement, and fast, and it provides results good enough for many real-world scenarios, especially in text classification.

Here’s a step-by-step explanation of how it works:

1. **Bayes' Theorem:** The algorithm is based on Bayes' theorem, which describes the probability of an event, based on prior knowledge of conditions that might be related to the event. Mathematically, it's expressed as:
   
   $$
   P(A|B) = \frac{P(B|A) \times P(A)}{P(B)}
   $$

   Here, $P(A|B)$ is the probability of hypothesis $A$ given the data $B$, $P(B|A)$ is the probability of the data given that the hypothesis is true, $P(A)$ is the probability of the hypothesis before seeing the data, and $P(B)$ is the probability of the data under all hypotheses.

2. **Naive Assumption:** The "naive" assumption is that all features (or inputs) are independent of each other given the class. This simplifies the computation, as the probability of observing the entire set of features is the product of the probabilities of observing each feature individually:
   
   $$
   P(B|A) = P(x_1|A) \times P(x_2|A) \times \dots \times P(x_n|A)
   $$
   
   This is where the algorithm gets its name, as assuming feature independence is often a naive approach since in reality, features can be dependent.

3. **Training the Model:** In the training phase, the algorithm calculates the probabilities of the features within each class from the training data.

4. **Classification:** For classification, it applies the Bayes' theorem using the naive independence assumption. It calculates the probability of each class given the input feature set and classifies the input to the class with the highest posterior probability.

Despite its simplicity, the Naive Bayes algorithm can outperform more sophisticated classification methods. Its performance depends heavily on the correct feature independence assumption. When features are not truly independent, the algorithm can still work well if the dependence structure between features doesn't diverge too much from the assumption, or if the dependencies are distributed evenly across classes, cancelling out the bias in the final decision.

Explain the trade-off between precision and recall in the context of binary classification. Provide an example scenario where optimizing one metric might come at the expense of the other. How can this trade-off be managed to make an informed decision when building a classification model?


> <font color="orange">**Note:**</font>
Present a practical example scenario, such as a medical test for a rare disease. Explain how optimizing one metric (e.g., precision) may result in fewer false alarms (fewer healthy patients incorrectly identified as having the disease) but could miss some true cases of the disease (lower recall).


In binary classification, precision and recall are metrics that evaluate the quality of a model. Precision measures the proportion of true positive results among all the instances classified as positive, while recall measures the proportion of true positive results among all the instances that should have been classified as positive.

**Precision** is defined as:
$$ \text{Precision} = \frac{\text{True Positives}}{\text{True Positives} + \text{False Positives}} $$

**Recall** is defined as:
$$ \text{Recall} = \frac{\text{True Positives}}{\text{True Positives} + \text{False Negatives}} $$

**The Trade-Off:**
- **High Precision, Lower Recall:** Optimizing for precision minimizes false positives but can result in missing true positives (because the model is more conservative and predicts positive only when it's highly certain). This can lower recall since some true cases are not identified.
- **High Recall, Lower Precision:** Optimizing for recall maximizes the number of true positives identified but can increase false positives (because the model predicts positive more liberally). This can reduce precision as the number of false alarms rises.

**Example Scenario in Medical Testing:**
Let’s consider a medical test for a rare but serious disease. If a medical test is designed to optimize precision, it will only identify a patient as having the disease if there's strong evidence. This would mean fewer healthy patients are incorrectly identified as having the disease, minimizing false alarms and unnecessary anxiety or invasive follow-up tests. However, this could result in a lower recall, missing some actual cases of the disease, which could be dangerous if early treatment is crucial.

On the other hand, if the test is optimized for recall, it will identify as many patients as possible who might have the disease, minimizing the chance of missing a true case. The downside is that more healthy patients will be incorrectly identified as having the disease (lower precision), leading to more false alarms and potentially unnecessary treatments or further testing.

**Managing the Trade-Off:**
To manage this trade-off, one might use a metric like the F1 score, which is the harmonic mean of precision and recall, or adjust the classification threshold based on the costs of false positives and false negatives. In the context of a rare disease, the cost of missing a true case (false negative) might be higher than the cost of additional testing due to a false alarm (false positive), so a higher recall might be prioritized. However, if the follow-up testing is expensive, risky, or has limited availability, one might instead prioritize precision to minimize the burden on the healthcare system and patients.

In practice, the decision often involves stakeholders (like medical professionals in our example) determining the acceptable balance between these metrics, considering the implications of false positives and false negatives on the patients and the healthcare system.

# Outlier Filteration

We'll identify potential outliers in the numerical data using statistical methods such as the Interquartile Range (IQR). After that, we can proceed with preprocessing and feature engineering, including adding second-order and interaction features and applying transformations such as logarithmic scaling on certain features. Let's start by detecting outliers.​

The outlier detection process identified 17,584 entries as outliers based on the Interquartile Range (IQR) method across the numerical columns. This is a significant portion of the dataset, and such a high number of outliers is not uncommon in large datasets, especially when considering that some features like `capital_gain` and `capital_loss` have a lot of zeros and then some very large values which can skew the IQR calculation.

Given the nature of this dataset and the fact that income levels can vary greatly due to factors such as hours worked or capital gains, it may not be appropriate to remove all these outliers. Instead, we should carefully consider the context of each feature and how it relates to the target variable income.

(We will back to this step and apply filteration and compare results.)

```python
# Function to detect outliers based on IQR
def detect_outliers(df, features):
    outlier_indices = []
    
    for c in features:
        Q1 = df[c].quantile(0.25)
        Q3 = df[c].quantile(0.75)
        IQR = Q3 - Q1
        outlier_step = 1.5 * IQR
        
        # Determine a list of indices of outliers for feature c
        outlier_list_col = df[(df[c] < Q1 - outlier_step) | (df[c] > Q3 + outlier_step)].index
        
        # Append the found outlier indices for c to the list of outlier indices 
        outlier_indices.extend(outlier_list_col)
    
    # Select observations containing more than 2 outliers
    outlier_indices = list(set(outlier_indices))
    return outlier_indices

# Detect outliers from numerical columns
numerical_cols = data.select_dtypes(include=['int64']).columns.tolist()
outliers = detect_outliers(data, numerical_cols)

# Display the outliers for review
outlier_data = data.loc[outliers]
print(len(outliers))
outlier_data.head()
```

# Grouping Sparse Classes

Now, Let's apply more Advanced Step, Grouping sparse classes in categorical features to reduce dimensionality.

```python
def group_sparse_classes(dataframe, threshold=0.01, apply_grouping=True):
    if apply_grouping:
        for col in dataframe.select_dtypes(include=['object']).columns:
            threshold_value = threshold * len(dataframe)
            value_counts = dataframe[col].value_counts()            
            # Identify values that are below the threshold
            sparse_classes = value_counts[value_counts < threshold_value].index
            # Replace the sparse classes with 'Other'
            dataframe[col] = dataframe[col].replace(sparse_classes, 'Other')
            
    return dataframe

apply_grouping = True  
data_encoded_feature_engineered = group_sparse_classes(data_encoded_feature_engineered, apply_grouping=apply_grouping)
```

# Feature Scaling

Three common types of feature scaling are:

- **Standard Scaling (Z-score normalization)**: This rescales the feature to have a mean (μ) of 0 and standard deviation (σ) of 1. The standard score of a sample $ x $ is calculated as $ z = \frac{(x - \mu)}{\sigma} $.

- **MinMax Scaling**: This rescales the feature to a fixed range, usually 0 to 1. The transformation is given by $ X_{std} = \frac{(X - X_{min})}{(X_{max} - X_{min})} $, and then $ X_{scaled} = X_{std} \times (max - min) + min $.

- **Robust Scaling**: This method scales features using statistics that are robust to outliers. It removes the median and scales the data according to the quantile range (defaults to IQR: Interquartile Range). The IQR is the range between the 1st quartile (25th quantile) and the 3rd quartile (75th quantile).

```python
from sklearn.preprocessing import StandardScaler, MinMaxScaler, RobustScaler

def scale_features(data, numerical_cols, scale_type='standard'):
    scaled_data = data.copy()
    
    if scale_type == 'standard':
        scaler = StandardScaler()
    elif scale_type == 'minmax':
        scaler = MinMaxScaler()
    elif scale_type == 'robust':
        scaler = RobustScaler()
    else:
        raise ValueError("`scale_type` should be 'standard', 'minmax', or 'robust'.")

    # Fit and transform the numerical features
    scaled_data[numerical_cols] = scaler.fit_transform(scaled_data[numerical_cols])
    
    return scaled_data

# Choose 'standard', 'minmax', or 'robust' for the scale_type parameter
data_encoded_feature_engineered_scaled = scale_features(
    data_encoded_feature_engineered,
    numerical_cols=['age', 'fnlwgt', 'education_ordinal', 'hours_per_week'], 
    scale_type='standard'
)
```

To provide results and conclusions for the 'KNN', 'LDA', 'QDA', and 'Logistic Regression' models, I will give a hypothetical summary based on typical outcomes from these models. Since the actual results depend on the specific dataset and experiment setup, the following is an illustrative example:

### K-Nearest Neighbors (KNN)
**What was done:** The KNN algorithm was applied to classify instances based on the majority vote of the nearest neighbors. The number of neighbors (k) was optimized based on cross-validation scores.

**What was found:** KNN performed moderately well on the dataset, suggesting that there is some local structure that KNN can exploit. However, its performance was sensitive to the scale of the data, and feature scaling was necessary to improve its accuracy. The model's performance was slightly lower on the validation set, indicating potential overfitting as k was increased.

### Linear Discriminant Analysis (LDA)
**What was done:** LDA was used to find a linear combination of features that best separate two or more classes of the dataset. LDA assumes that different classes generate data based on Gaussian distributions with different means but the same covariance matrix.

**What was found:** LDA offered a good balance between bias and variance, providing a robust model that performed well on both the training and validation datasets. The assumption of common covariance seemed to hold reasonably well for the given data, and LDA served as a good baseline model.

### Quadratic Discriminant Analysis (QDA)
**What was done:** QDA, similar to LDA, was employed but without the assumption of equal covariance matrices across classes. This allows for a quadratic decision boundary.

**What was found:** QDA captured more complex structures in the data compared to LDA. While it outperformed LDA in areas where the class distributions have significantly different covariances, it was also more prone to overfitting, especially in higher dimensions or with limited training data.

### Logistic Regression
**What was done:** Logistic regression was used to model the probability that an instance belongs to a particular class. Being a parametric model, it assumes a linear relationship between the log-odds of the probabilities and the input features.

**What was found:** Logistic regression provided a strong baseline with the advantage of interpretability of the model coefficients. It performed well in terms of calibration of predicted probabilities. However, its linear nature meant that it did not capture interactions or non-linear decision boundaries as effectively as some non-linear models.

### Conclusion
Each model has its strengths and limitations, often dictated by the underlying assumptions about the data. KNN can capture complex patterns but may suffer from high variance. LDA is robust and efficient for data that meets its assumptions, while QDA can adapt to more varied class distributions at the cost of potentially overfitting. Logistic Regression offers interpretability and probabilistic outputs, making it a strong candidate for binary classification problems.

Ultimately, the choice of model should be guided by the problem specifics, data characteristics, and the trade-off between complexity and interpretability. Cross-validation and ROC-AUC plots can help in assessing model performance and ensuring that the conclusions are generalizable to unseen data.
